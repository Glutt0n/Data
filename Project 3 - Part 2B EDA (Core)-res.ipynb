# Exploratory Data Analysis

```python
# imports
import pandas as pd
import json
```

```python
# Loading in data
tmdb_2000=pd.read_csv('Desktop/Project_3/Project_3final_tmdb_data_2000.csv.gz')
tmdb_2001=pd.read_csv('Desktop/Project_3/Project_3final_tmdb_data_2001.csv.gz')
display(tmdb_2000.head())
tmdb_2001.head()
```

```python
# concatenatingi data into a single dataframe
movies=pd.concat([tmdb_2000,tmdb_2001])
movies.head()
```

```python
# How many movies had at least some valid financial information (values > 0 for 
# budget and revenue)?
money_filter=(movies['budget']>0)|(movies['revenue']>0)
movies=movies[money_filter].copy()
print(movies['title'].count()) #counting all titles after filtering
```

```python
# How many movies are there in each of the certification categories (G/PG/PG-13/R)?
movies['certification'].value_counts()
```

```python
# What is the average revenue per certification category?
movies[['revenue','certification']].groupby('certification').mean().round()
```

```python
# What is the average budget per certification category?
movies[['budget','certification']].groupby('certification').mean().round()
```

```python
# Saving and compressing tmdb data
movies.to_csv('tmdb_results_combined.csv.gz',compression='gzip',index=False)
```
